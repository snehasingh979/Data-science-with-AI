```python
def final_salary(basic_salary, bonus, tax_percent):
    return (basic_salary + bonus) - ((basic_salary + bonus) * tax_percent / 100)

salary = float(input("Enter basic salary: "))
bonus = float(input("Enter bonus: "))
tax = float(input("Enter tax percentage: "))

print("Final Salary =", final_salary(salary, bonus, tax))
```
