Module 4-Selecting Data

```python
import pandas as pd
df=pd.read_csv("//content/Employee.csv")



print(df["Gender"]) # Selecting a Single Column
df[["Education", "City"]] # selecting multiple columns

#Selecting Rows using Index

print(df.iloc[0]) # first row
print(df.iloc[0:5]) # multiple rows
print(df.iloc[-5:]) # last five rows

## Selecting Rows and Columns using loc[]

#loc[] uses row labels (index names) and column names.
# syntax -> df.loc[row_label, column_name]
print(df.loc[0]) # first row
print(df.loc[0:5]) # multiple rows
print(df.loc[-5:]) # last five rows

#Selecting Rows and Columns using iloc[]

#iloc[] uses integer positions.
#syntax -> df.iloc[row_index, column_index]

print(df.iloc[0]) #first row
print(df.iloc[0:5]) #multiple rows
print(df.iloc[-5:]) #last five rows

# Boolean Indexing

#Boolean Indexing returns rows where the condition is True.

# Age greater than 30
df[df["Age"] > 30]

# Age less than 30
df[df["Age"] < 30]
# city is banglore
df[df["City"] == "Bangalore"]

# education is master
df[df["Education"] == "Masters"]

# Filtering Data

df[df["City"] == "Bangalore"] # single condition
df[(df["City"] == "Pune") & # multiple conditions(AND) (
   (df["Age"] > 30)]

df[(df["City"] == "Pune") | # multiple conditions(OR)
   (df["Age"] > 30)]

df[~(df["City"] == "New Delhi")] # NOT condition (changed from Department == Sales)

df[df["City"].isin(["Bangalore", "Pune"])] # using isin()

df[df["Age"].between(25, 35)] # using(between)

df[df["Education"].str.contains("Bachelor", case=False)] # using str.contains()
# using query
df.query("Age > 30")
df.query("City == 'Pune'")
df.query("Age > 30 and ExperienceInCurrentDomain > 3")

df.nlargest(5, "Age") # using nlargest()
df.nsmallest(5, "Age") # using nsmallest()
```

Module 5- Missing value in pandas

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Aman", "Riya", "Karan", "Priya", np.nan],
    "Age": [22, np.nan, 25, 28, 24],
    "City": ["Delhi", "Mumbai", np.nan, "Lucknow", "Delhi"],
    "Salary": [50000, 60000, np.nan, 75000, 55000]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)
print(df.isnull()) #. used to identify missing value
print(df.isnull().sum()) # count missing value
print(df.notnull()) #. used to identify missing value
# fill with a constant
filled = df.fillna("Unknown")
print(filled)
# fill numeric coulmns
df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)
# fill multiple columns
df.fillna({
    "Name": "Not Available",
    "City": "Unknown",
    "Salary": 0
}, inplace=True)

print(df)

df.dropna() # removes missing values
# df.replace(old_value, new_value) # used to replace speific values (not limited to missing values)
```

Module-8 String function

```python
import pandas as pd

# Sample Dataset
df = pd.DataFrame({
    "Name": ["Amit", "Priya", "Rahul", "Neha", "David"]
})

print(df)

# .str.lower() -> converts all text to lowercase
df["Lower"] = df["Name"].str.lower()

print(df)

# .str.upper() -> Converts all text to uppercase.
df["Upper"] = df["Name"].str.upper()

print(df)

# .str.contains() -> Checks whether a string contains a specific word or character.
df["Contains_a"] = df["Name"].str.contains("a", case=False)

print(df)

# .str.split() -> Splits a string into multiple parts.
df["Split"] = df["Name"].str.split()

print(df)

# .str.replace() -> Replaces a word or character with another.
df["Replaced"] = df["Name"].str.replace("a", "A")

print(df)

# .str.strip() -> Removes leading and trailing spaces.
df["Stripped"] = df["Name"].str.strip()

print(df)
```

Module-9-Group By

topics

1.grouping
2.mean()
3.sum()
4.count()
5.agg()

```python
import pandas as pd

data = {
    "Employee": ["Aman", "Riya", "Rahul", "Priya", "Karan", "Neha"],
    "Department": ["HR", "IT", "IT", "HR", "Sales", "Sales"],
    "Salary": [40000, 60000, 55000, 45000, 50000, 52000]
}

df = pd.DataFrame(data)

print(df.groupby("Department")) # group by object
print(df.groupby("Department")["Salary"].mean()) # mean of 'Salary' for each group
print(df.groupby("Department")["Salary"].sum()) # sum of 'Salary' for each group
print(df.groupby("Department").count()) # count of each group
print(df.groupby("Department").agg({"Salary": ["mean", "sum"]})) # aggregate
```

Module 10- Merge & Join

1.merge

```python
import pandas as pd
# Employee Data
employees = pd.DataFrame({
    "Emp_ID": [101, 102, 103, 104],
    "Name": ["Amit", "Neha", "Rahul", "Priya"]
})
# Salary Data
salary = pd.DataFrame({
    "Emp_ID": [101, 102, 103, 104],
    "Salary": [45000, 52000, 48000, 60000]
})
# Merge DataFrames
result = pd.merge(employees, salary, on="Emp_ID")
print(result)
```

Module 11 – Pivot Tables

is used to reshape data by converting unique values from one column into multiple columns. It does not perform aggregation, so
duplicate combinations are not allowed.
keyboard_arrow_down

```python
import pandas as pd
# Dataset
data = {
    "Employee": ["A", "A", "B", "B"],
    "Month": ["Jan", "Feb", "Jan", "Feb"],
    "Sales": [5000, 7000, 6000, 8000]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n")
print(df)
print("\nPivot Table:\n")
pivot_df = df.pivot(
    index="Employee",
    columns="Month",
    values="Sales"
)
print(pivot_df)
```

2. pivot_table()

```python
import pandas as pd
# Dataset
data = {
    "Department": ["IT", "IT", "HR", "HR", "IT"],
    "Gender": ["Male", "Female", "Male", "Female", "Male"],
    "Salary": [50000, 60000, 45000, 48000, 55000]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n")
print(df)
print("\nPivot Table using Mean:\n")
pivot_table_df = df.pivot_table(
    values="Salary",
    index="Department",
    columns="Gender",
    aggfunc="mean"
)
print(pivot_table_df)
```

3. crosstab()

```python
import pandas as pd
# Dataset
data = {
    "Department": ["IT", "IT", "HR", "HR", "IT", "Sales", "Sales"],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male"]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n")
print(df)
print("\nCrossTab:\n")
cross = pd.crosstab(
    df["Department"],
    df["Gender"]
)
print(cross)
```
