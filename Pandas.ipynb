Main Data Structure in Pandas

2.Data frame A dataframe is two-dimiensional labeled array consisting row and columns

1.Creating series

```python
import pandas as np
marks = np.Series([85,90,78,92,88])
print(marks)
```

creating a series of string

```python
import pandas as np
fruits = np.Series(['apple','banana','orange','mango'])
print(fruits)
```

creating a series with custom index

```python
import pandas as np
marks = np.Series([85,90,78,92,88],index=['ram','shyam','hari','gita','sita'])
print(marks)
```

creating a series form a dictionary

```python
import pandas as pd
student = pd.Series({"Name":"Aman","Age":20,"city":"delhi"})
print(student)
```

Access value in  a series

```python
import pandas as pd
marks = pd.Series([85,90,78,92,88],index=['ram','shyam','hari','gita','sita'])
print(marks[2])
```

Access value using  custom index

```python
import pandas as pd
marks = pd.Series([85,90,78,92,88],index=['ram','shyam','hari','gita','sita'])
print(marks['hari'])
```

Print Series information

```python
import pandas as pd
marks = pd.Series([85,90,78,92,88],index=['ram','shyam','hari','gita','sita'])
print(marks.dtype)
```

2.creating Data frame

dataframe is a two dimensional labeled table consisting of row and columne A  data frame is used to store, organzinng ,manipulate,and analyzer tabular data efficienctly

syntax

important pandas as pd df=pd.data frame(data)

```python
import pandas as pd
student={
    "Name":["Aman","Ravi","priya","neha"],
    "Age":[20,21,19,22],
    "city":["delhi","lucknow","ghaziabad","noida"]
}
df=pd.DataFrame(student)
print(df)
```

create dataframe with customer index

```python
import pandas as pd
student={
    "Name":["Aman","Ravi","priya","neha"],
    "Marks":[85,90,78,92],
    "city":["delhi","lucknow","ghaziabad","noida"]
}
df=pd.DataFrame(student,index=["s1","s2","s3","s4"])
print(df)
```

create dataframe from a list of list

```python
import pandas as pd
data=[
    ["Aman",85,"delhi"],
    ["Ravi",90,"lucknow"],
    ["priya",78,"ghaziabad"],
    ["neha",92,"noida"]
]
df=pd.DataFrame(data,columns=["Name","Marks","city"])
print(df)
```

create dataframe from a list of Dictionaries

```python
import pandas as pd
data=[
    {"Name":"Aman","Age":20},
    {"Name":"Ravi","Age":21},
    {"Name":"priya","Age":19},
    {"Name":"neha","Age":22}
]
df=pd.DataFrame(data)
print(df)
```
