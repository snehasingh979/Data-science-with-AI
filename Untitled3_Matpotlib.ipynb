```python
# First Graph using Matplotlib

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Create Figure
plt.figure(figsize=(8, 5))

# Plot Line Graph
plt.plot(
    x,
    y,
    marker='o',
    linestyle='-',
    linewidth=2,
    color='blue',
    label='Sales'
)

# Title and Labels
plt.title("First Line Graph", fontsize=16)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)


# Save the graph
plt.savefig("first_graph.png", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.jpg", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.pdf", bbox_inches="tight")

# Display the graph
plt.show()

# Close the figure
plt.close()
```

Module 3- Bar charts

dataset employes salary dataset topics

1.bar()

2.barch()

3.color

4.edgecolor

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load Dataset
df = pd.read_csv("/content/employee_salary.csv")

# =====================================================
# 1. bar()
# =====================================================


salary_by_age = df.groupby("age")["income"].mean()

plt.figure(figsize=(10,6))
plt.bar(salary_by_age.index.astype(str), salary_by_age.values)

plt.title("Basic Bar Chart (Mean Income by Age)")
plt.xlabel("Age")
plt.ylabel("Mean Income")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# =====================================================
# 2. barh()
# =====================================================

plt.figure(figsize=(10,6))
plt.barh(salary_by_age.index.astype(str), salary_by_age.values)

plt.title("Horizontal Bar Chart (Mean Income by Age)")
plt.xlabel("Mean Income")
plt.ylabel("Age")
plt.tight_layout()
plt.show()

# =====================================================
# 3. width
# =====================================================

plt.figure(figsize=(10,6))
plt.bar(
    salary_by_age.index.astype(str),
    salary_by_age.values,
    width=0.4
)

plt.title("Bar Width Example (Mean Income by Age)")
plt.xlabel("Age")
plt.ylabel("Mean Income")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# =====================================================
# 4. color
# =====================================================

plt.figure(figsize=(10,6))

plt.bar(
    salary_by_age.index.astype(str),
    salary_by_age.values,
    color="orange"
)

plt.title("Bar Color Example (Mean Income by Age)")
plt.xlabel("Age")
plt.ylabel("Mean Income")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# =====================================================
# 5. edgecolor
# =====================================================

plt.figure(figsize=(10,6))

plt.bar(
    salary_by_age.index.astype(str),
    salary_by_age.values,
    color="skyblue",
    edgecolor="black"
)

plt.title("Edge Color Example (Mean Income by Age)")
plt.xlabel("Age")
plt.ylabel("Mean Income")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

A histogram is a graphical represenation used to displays the frequency distribution of continuous numerical data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset

df = pd.read_csv("/content/house_prices.csv")
print("DataFrame Columns:", df.columns)

# =====================================================
# 1. Basic Histogram
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(df["Price (in rupees)"]) # A histogram groups numerical data into intervals (bins) and displays the frequency.

plt.title("Basic Histogram")

plt.show()

# =====================================================
# 2. bins
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(                         # bins decides how many intervals (groups) the data is divided into.
    df["Price (in rupees)"],
    bins=20
)

plt.title("Histogram using Bins")

plt.show()

# =====================================================
# 3. density
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"], # Instead of showing frequency, it shows the Probability Density.
    bins=20,
    density=True
)

plt.title("Density Histogram")

plt.show()

# =====================================================
# 4. cumulative histogram
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"],  # Shows the cumulative frequency.
    bins=20,
    cumulative=True
)

plt.title("Cumulative Histogram")

plt.show()

# =====================================================
# 5. transparency (alpha)
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"], # Transparency is controlled using the alpha parameter.
    bins=20,
    alpha=0.5
)

plt.title("Histogram using Transparency")

plt.show()
```

scatter plot

A scatter plot is used to vislization the relationship between two numerical variables

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("/content/Iris.csv")

# =====================================================
# 1. Basic scatter()
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"]
)

plt.title("Basic Scatter Plot")

plt.show()

# =====================================================
# 2. size (s). # The s parameter controls the size of the markers.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    s=100
)

plt.title("Scatter Plot with Size")

plt.show()

# =====================================================
# 3. color. # Changes the color of markers.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    color="red"
)

plt.title("Scatter Plot with Color")

plt.show()

# =====================================================
# 4. alpha. # Controls the transparency.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    alpha=0.5
)

plt.title("Scatter Plot with Transparency")

plt.show()

# =====================================================
# 5. marker. # Changes the shape of the points.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    marker="^"
)

plt.title("Scatter Plot with Marker")

plt.show()

# =====================================================
# 6. colormap (cmap) # A colormap assigns colors to points based on numeric values in c.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    c=df["PetalWidthCm"],
    cmap="viridis",
    s=80
)

plt.colorbar(label="Petal Width")

plt.title("Scatter Plot with Colormap")

plt.show()
```

Pie chart

A pie chart is used to displays the proportion or percentage of each category relative to the whole

```python
# =============================================
# Module 6 - Pie Charts
# Dataset : titanic.csv
# Perform All Operations Individually
# =============================================

import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("/content/titanic.csv")

# Survival Count
survival = df["Survived"].value_counts()

labels = ["Not Survived", "Survived"]

# =====================================================
# 1. Basic pie()
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(survival.values)

plt.title("Basic Pie Chart")

plt.show()

# =====================================================
# 2. labels # Adds labels to each slice.
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels
)

plt.title("Pie Chart with Labels")

plt.show()

# =====================================================
# 3. autopct # Displays percentage values on each slice.
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels,
    autopct="%1.1f%%"
)

plt.title("Pie Chart with Percentage")

plt.show()

# =====================================================
# 4. explode # explode is used to separate one or more slices from the pie chart.
# =====================================================

explode = [0, 0.15]

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels,
    explode=explode,
    autopct="%1.1f%%"
)

plt.title("Pie Chart using Explode")

plt.show()

# =====================================================
# 5. shadow # Adds a shadow behind the pie chart.
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels,
    shadow=True,
    autopct="%1.1f%%"
)

plt.title("Pie Chart with Shadow")

plt.show()

# =====================================================
# 6. startangle # Rotates the starting position of the pie chart.

# Default starting angle = 0°
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels,
    startangle=90,
    autopct="%1.1f%%"
)

plt.title("Pie Chart using Start Angle")

plt.show()
```

Module 7- Box plot

Dataset - employes salary dataset topic

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("/content/employee_salary.csv")

# =====================================================
# 1. Basic boxplot() # boxplot() is a Matplotlib function used to create a box plot.
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(df["income"])

plt.title("Basic Box Plot")

plt.show()

# =====================================================
# 2. Outliers # An Outlier is a value that is much larger or much smaller than the other data values.
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(
    df["income"],
    showfliers=True
)

plt.title("Box Plot with Outliers")

plt.show()

# =====================================================
# 3. notch #The notch adds a small indentation around the median.

# It helps compare the medians of different datasets.
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(
    df["income"],
    notch=True
)

plt.title("Notched Box Plot")

plt.show()

# =====================================================
# 4. patch_artist # patch_artist=True allows you to fill the box with color.
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(
    df["income"],
    patch_artist=True,
    boxprops=dict(facecolor="skyblue")
)

plt.title("Colored Box Plot")

plt.show()
```
