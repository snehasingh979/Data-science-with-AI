```python
# First Graph using Matplotlib

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Create Figure
plt.figure(figsize=(8, 5))

# Plot Line Graph
plt.plot(
    x,
    y,
    marker='o',
    linestyle='-',
    linewidth=2,
    color='blue',
    label='Sales'
)

# Title and Labels
plt.title("First Line Graph", fontsize=16)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)


# Save the graph
plt.savefig("first_graph.png", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.jpg", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.pdf", bbox_inches="tight")

# Display the graph
plt.show()

# Close the figure
plt.close()
```

Module 3- Bar charts

dataset employes salary dataset topics

4.edgecolor

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load Dataset
df = pd.read_csv("/content/employee_salary.csv")

# =====================================================
# 1. bar()
# =====================================================


salary_by_age = df.groupby("age")["income"].mean()

plt.figure(figsize=(10,6))
plt.bar(salary_by_age.index.astype(str), salary_by_age.values)

plt.title("Basic Bar Chart (Mean Income by Age)")
plt.xlabel("Age")
plt.ylabel("Mean Income")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# =====================================================
# 2. barh()
# =====================================================

plt.figure(figsize=(10,6))
plt.barh(salary_by_age.index.astype(str), salary_by_age.values)

plt.title("Horizontal Bar Chart (Mean Income by Age)")
plt.xlabel("Mean Income")
plt.ylabel("Age")
plt.tight_layout()
plt.show()

# =====================================================
# 3. width
# =====================================================

plt.figure(figsize=(10,6))
plt.bar(
    salary_by_age.index.astype(str),
    salary_by_age.values,
    width=0.4
)

plt.title("Bar Width Example (Mean Income by Age)")
plt.xlabel("Age")
plt.ylabel("Mean Income")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# =====================================================
# 4. color
# =====================================================

plt.figure(figsize=(10,6))

plt.bar(
    salary_by_age.index.astype(str),
    salary_by_age.values,
    color="orange"
)

plt.title("Bar Color Example (Mean Income by Age)")
plt.xlabel("Age")
plt.ylabel("Mean Income")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# =====================================================
# 5. edgecolor
# =====================================================

plt.figure(figsize=(10,6))

plt.bar(
    salary_by_age.index.astype(str),
    salary_by_age.values,
    color="skyblue",
    edgecolor="black"
)

plt.title("Edge Color Example (Mean Income by Age)")
plt.xlabel("Age")
plt.ylabel("Mean Income")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Employee count Bar chart

objective

Display the total number of employees in each department

```python
employee_count_by_age = df['age'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(employee_count_by_age.index.astype(str), employee_count_by_age.values)

plt.title('Total Number of Employees by Age')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Practical 3-Experiance comparsion bar chart

objective

compare the average years of experience across department

```python
average_experience_by_age = df.groupby('age')['experience'].mean()

plt.figure(figsize=(12, 6))
plt.bar(average_experience_by_age.index.astype(str), average_experience_by_age.values)

plt.title('Average Years of Experience by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Years of Experience')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
