1.Measure of Central tendency

measure of central tendency help us the middle or "typical" value of a collection of data

1. Mean

average value of all values formaula mean=

```python
import pandas as pd
import numpy as np


data = {
    "Employee": [
        "A", "B", "C", "D", "E",
        "F", "G", "H", "I", "J"
    ],
    "Age": [
        22, 25, 28, 30, 35,
        40, 32, 29, 45, 38
    ],
    "Salary": [
        25000, 32000, 40000, 42000, 50000,
        62000, 47000, 43000, 75000, 68000
    ],
    "Experience": [
        1, 2, 4, 5, 8,
        12, 6, 5, 18, 15
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Employee Dataset")
print(df)

# =====================================
# Mean of All Numerical Columns
# =====================================

print("\nMean of Numerical Columns")
print(df.mean(numeric_only=True))

# =====================================
# Mean of Individual Columns
# =====================================

print("\nAverage Age:", df["Age"].mean())
print("Average Salary:", df["Salary"].mean())
print("Average Experience:", df["Experience"].mean())
```

 2. Mode

```python
import pandas as pd
import numpy as np


data = {
    "Employee": [
        "A", "B", "C", "D", "E",
        "F", "G", "H", "I", "J"
    ],
    "Age": [
        22, 25, 28, 30, 35,
        40, 32, 29, 45, 38
    ],
    "Salary": [
        25000, 32000, 40000, 42000, 50000,
        62000, 47000, 43000, 75000, 68000
    ],
    "Experience": [
        1, 2, 4, 5, 8,
        12, 6, 5, 18, 15
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Employee Dataset")
print(df)

# ============================================
# Calculate Mode
# ============================================

print("\nMode of Each Column")
print(df.mode())

# ============================================
# Mode of Individual Columns
# ============================================

print("\nMode of Age")
print(df["Age"].mode())

print("\nMode of Salary")
print(df["Salary"].mode())

print("\nMode of Experience")
print(df["Experience"].mode())

# ============================================
# First Mode Value (if multiple modes exist)
# ============================================

print("\nFirst Mode of Experience:", df["Experience"].mode()[0])
```

3-Median

The Median is the middle value after sorting the data.

If there are:

Odd observations --> Middle value
Even observation --> Average of two middle class

```python
import pandas as pd
import numpy as np


data = {
    "Employee": [
        "A", "B", "C", "D", "E",
        "F", "G", "H", "I", "J"
    ],
    "Age": [
        22, 25, 28, 30, 35,
        40, 32, 29, 45, 38
    ],
    "Salary": [
        25000, 32000, 40000, 42000, 50000,
        62000, 47000, 43000, 75000, 68000
    ],
    "Experience": [
        1, 2, 4, 5, 8,
        12, 6, 5, 18, 15
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("=" * 50)
print("Employee Dataset")
print("=" * 50)
print(df)

# Median of Individual Columns

print("\nMedian of Age:")
print(df["Age"].median())

print("\nMedian of Salary:")
print(df["Salary"].median())

print("\nMedian of Experience:")
print(df["Experience"].median())

# Median of All Numerical Columns

print("\n" + "=" * 50)
print("Median of All Numerical Columns")
print("=" * 50)

print(df.median(numeric_only=True))
```

2- Measure of Dispersion
Measure of dispersion describe the variability or spread of data in a dataset.

1-Range

The simplest measure of dispersion.

Formula

Range= Maximum value - Minimum value

```python
import numpy as np
data=[12,18,20,25,30]
print("Range=",max(data)-min(data))
```

2- Interquartile Range(IQR)

IQR measures the spread of the middle 50% of the data.

It ignores extreme values.

Formula

IQR=Q3-Q1 Where
Q1= First Quartile(25%)
Q3- Third Quartile(75%)

```python
import numpy as np
data=[2,4,6,8,10,12,14,16]
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
print("IQR=", Q3-Q1)
print("Q1=",Q1)
print("Q3=",Q3)
```

3-Quartile Deviation (Semi-IQR)

Quartile Deviation is half of IQR

Formula

QD=2Q3-Q1

```python
import numpy as np
data=[2,4,6,8,10,12,14,16]
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
QD=(Q3-Q1)/2
print("Quartile Deviation=",QD)
```

5- Variance
Variance measures the average squared deviation from the mean. A small variance means data points are close to the mean,while a large variance means they are widely spread.

Formula
var(X)=E(X2)-[E(X)]2

There are two types:

Population Variance
σ2= ∑(x-μ)2/2

Sample Variance
s2=∑(x-x^-)2/n-1

6- Standard Deviation
Standard deviation is the square root of variance.It is most widely used measure of dispersion because it is expressed in the same units as the original data.

Population Standard Deviation
σ=vσ2

```python
import numpy as np
data=[2,4,6,8]
print("Population SD=",np.std(data))
print("Sample SD=",np.std(data,ddof=1))
```
